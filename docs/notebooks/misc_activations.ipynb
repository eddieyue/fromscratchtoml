```python
from fromscratchtoml.neural_network import Activations
from fromscratchtoml.toolbox import binary_visualize

import numpy as np

# set this to avoid exponential notations
np.set_printoptions(suppress=True, formatter={'float': '{: 0.8f}'.format})
%matplotlib inline
```

# Visualising activation funtions

```python
# little hack to get all the available activation functions
for k, v in Activations.__dict__.iteritems():
    if "method" in str(v):
        print(k)
```

## Linear Activation

```python
X = np.linspace(-5,5,20)
Y = Activations.linear(X)
XY = np.stack([X, Y], axis=1)
print(repr(Y))
binary_visualize(XY)
```
